**batting Summary**

```python
with open("project/t20_wc_batting_summary.json") as j_b:
    batting=json.load(j_b)
    records_batting=[]
    for rec in batting:
        records_batting.extend(rec['battingSummary'])
```

```python
batting
```

```python
len(records_batting)
```

```python
data_batting=pd.DataFrame(records_batting)
```

```python
pd.set_option('display.max_columns',700)
pd.set_option('display.max_rows',700)
data_batting
```

```python
data_batting.head(10)
```

```python
data_batting['Out or Not out']=data_batting['dismissal'].apply(lambda x: 'out' if len(x)>0 else "Not out")
```

```python
data_batting.head(10)
```

```python
data_batting.drop(columns=['dismissal'],inplace=True)
```

```python
data_batting.head(10)
```

```python
data_batting['batsmanName']=data_batting['batsmanName'].apply(lambda x: x.replace('â€',""))
data_batting['batsmanName']=data_batting['batsmanName'].apply(lambda x: x.replace('\xa0',""))
```

```python
data_batting.head(30)
```

```python
match_id_dict= {}
for index,row in data_match_summary.iterrows():
    key1=row['team1']+" Vs "+row['team2']
    key2=row['team2']+" Vs "+row['team1']
    match_id_dict[key1]=row['Match_id']
    match_id_dict[key2]=row['Match_id']

    
```

```python
match_id_dict
```

```python
data_batting['match_id']=data_batting['match'].map(match_id_dict)
```

```python
data_batting.head(5)
```
